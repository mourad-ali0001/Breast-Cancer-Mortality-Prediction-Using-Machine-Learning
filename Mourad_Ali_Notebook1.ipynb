Code Leveraged from Code Reuse Session 1,Block 17

```python
#I built a function that detects outliers in any variable of my choice based on the interquartile range
def find_outliers_IQR(data):

  q1=data.quantile(0.25)

  q3=data.quantile(0.75)

  IQR=q3-q1

  outliers = data[((data<(q1-1.75*IQR))|(data>(q3+1.75*IQR)))]

  return outliers
```

Code Leveraged from Code Reuse Session 1,Block 19


```python
#I called my built function to display the detected outliers in my chosen variables (Survival_Months, Age and Tumor_Size).
outliers = find_outliers_IQR(data['Survival_Months'])

print("number of outliers: "+ str(len(outliers)))

outliers
```
